```python
import pandas as pd
import requests
import os
from openai import OpenAI
from bs4 import BeautifulSoup
client = OpenAI()
```

```python
df = pd.read_csv('../free_api_urls.csv')

df['Doc_URL']
```

```python
for index, url in df['Doc_URL'].items():
    
    try:
        res = requests.get(url)
    except:
        continue
        
    if res.status_code == 200:
        
        soup = BeautifulSoup(res.text, 'html.parser')
        text_content = soup.get_text(separator=' ', strip=True)

        tokens = text_content.split()
        token_limit = 117500

        limited_tokens = tokens[:token_limit]
        limited_text = ' '.join(limited_tokens)

        html_content = limited_text

        try:
            completion = client.chat.completions.create(
              model="gpt-4-turbo",
              messages=[
                {"role": "system", "content": "You are a helpful assistant skilled in analyzing text content to recognize API Documentations."},
                {"role": "user", "content": "Determine if the following text content is an API documentation that contains endpoints. If your conclusion is that it is an API documentation, only output 1. If there is no such useful information, only output 0."},
                {"role": "user", "content": "This API allows users to retrieve data seamlessly. For more information, visit our documentation"},
                {"role": "system", "content": "0"},
                {"role": "user", "content": "'LINK (with RDF option) Name link â\x80\x93 find related entries by using database cross-references URL form https://rest.kegg.jp/link/<target_db>/<source_db>[/<option>] <target_db> = <database> <source_db> = <database> <database> = drug | atc | jtc <option> = turtle | n-triple https://rest.kegg.jp/link/<target_db>/<dbentries>[/<option>] <dbentries> = KEGG database entries of the following <database> <database> = drug | atc | jtc <option> = turtle | n-triple Description This operation allows retrieval of cross-references within a few selected databases. The first form allows retrieval of database to database cross-references, while the second form allows retrieval for a selected number of entries. Examples /link/atc/D01441/n-triple ATC code of a given KEGG DRUG entry in N-Triples format /link/jtc/D01441/turtle Japanese therapeutic category of a given KEGG DRUG entry in Turtle format Last updated: May 26, 2023 « Back'"},
                {"role": "system", "content": "1"},
                {"role": "system", "content": html_content}
              ],
              temperature=0,
            )

            print(url)
            print(completion.choices[0].message.content)
            print()
            
        except Exception as e:
            print(e)
            continue
        
        if completion.choices[0].message.content == "1":
            file_name = str(index) + ".html"
            with open("../free_api_docs/" + file_name, 'w', encoding='utf-8') as file:
                file.write(res.text)
```

```python
folder_path = '../free_api_docs'

files = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

file_count = len(files)

print(f"There are {file_count} files in the directory.")
```
